# Dream House ##

````python
                                            ## DREAM HOUSE ##
    
##   Initials of the variable consisting time period have been abbreviated as follows for the simplcity of reading   ##
    # a_  --> annual eg. annual salary or annual returns of investment has been abbreviated as a_salary or a_roi
    # m_  --> monthly eg. monthly salary or monthly savings has been abbreviated as m_salary or m_savings
    # sa_ --> semi annually eg. semi annual raise in salary has been abbreviated as sa_salary_raise
    # perc -> percentage
    
    
    
#input users annual salary as integer    
a_salary = int(input("What's your annual salary? "))


# take the input for monthly saving percentage in salary in string format 
# if there is any % sign in it we  remove it using string split funtion.
# convert the string to integer value and used in while loop to make sure valid value is passed to the program.
# keep prompting to the user to enter the valid value 

m_save_perc_str  =  input("What percent of your salary will you save? ")
m_save_perc_str  =  m_save_perc_str.split('%')
m_save_perc      =  int(m_save_perc_str[0])

while (m_save_perc > 100) or (m_save_perc < 0):
    m_save_perc_str  =  input("What percent of your salary will you save? ")
    m_save_perc_str  =  m_save_perc_str.split('%')
    m_save_perc      =  int(m_save_perc_str[0])

    
    
# input price of the house and calculate downpayment.
house_price = int(input('How much does your dream home cost? '))
downpayment = house_price / 4


#initialise the variable to use it later for keeping the accumulated value of salary after every iteration of month. 
savings_total = 0

# calculating monthly salary
m_salary      = a_salary / 12

#initialise the variable to use it later as counter
months        = 0

#keep adding the monthly savings for every month till it becomes greater than downpayment and capture number of months.
while savings_total < downpayment:
    savings_total  += m_salary * ( m_save_perc / 100)
    months         += 1
    
print()
print('You would have to work for %d months to afford the downpayment for your dream house.'%months)
print('%d months or %d years and %d months.'%(months,months//12,months%12))
````

# Build Some Savings##

````python
                                    #BUILD  SOME SAVINGS##
    
##   Initials of the variable consisting time period have been abbreviated as follows for the simplcity of reading   ##
    # a_  --> annual eg. annual salary or annual returns of investment has been abbreviated as a_salary or a_roi
    # m_  --> monthly eg. monthly salary or savings has been abbreviated as m_salary or m_savings
    # sa_ --> semi annually eg. semi annual raise in salary has been abbreviated as sa_salary_raise
    # perc -> percentage


a_salary = float(input("What's your annual salary?"))
                                                      
# saving the annualy salary for using it later 
a_salary_base=a_salary 



# take the input for monthly saving percentage in salary in string format 
# if there is any % sign in it we  remove it using string split funtion.
# convert the string to integer value and used in while loop to make sure valid value is passed to the program.
# keep prompting to the user to enter the valid value 

m_save_perc_str  =  input("What percent of your salary will you save? ")
m_save_perc_str  =  m_save_perc_str.split('%')
m_save_perc      =  int(m_save_perc_str[0])

while (m_save_perc > 100) or (m_save_perc < 0):
    m_save_perc_str  =  input("What percent of your salary will you save? ")
    m_save_perc_str  =  m_save_perc_str.split('%')
    m_save_perc      =  int(m_save_perc_str[0])

#calculating downpayment value from the house price given by user.
house_price = int(input('How much does your dream home cost? '))
downpayment = float(house_price/4)


# take the input for semi annual raise percentage in salary in string format 
# if there is any % sign in it we  remove it using string split funtion.
# convert the string to integer value and used in while loop to make sure valid value is passed to the program.
# keep prompting to the user to enter the valid value 

sa_salary_raise_str  =  input("What's the semi-annual raise percentage? ")
sa_salary_raise_str  =  sa_salary_raise_str.split('%')
sa_salary_raise      =  int(sa_salary_raise_str[0])

while (sa_salary_raise > 100) or (sa_salary_raise < 0):
    sa_salary_raise_str  =  input("What's the semi-annual raise percentage? ")
    sa_salary_raise_str  =  sa_salary_raise_str.split('%')
    sa_salary_raise      =  int(sa_salary_raise_str[0])    

    
# take the input for annual return of investment percentage in salary in string format 
# if there is any % sign in it we  remove it using string split funtion.
# convert the string to integer value and used in while loop to make sure valid value is passed to the program.
# keep prompting to the user to enter the valid value 

a_roi_str  =  input("What's the investment annual return? ")
a_roi_str  =  a_roi_str.split('%')
a_roi      =  int(a_roi_str[0])

while (a_roi > 100) or (a_roi <0):
    a_roi_str  =  input("What's the investment annual return? ")
    a_roi_str  =  a_roi_str.split('%')
    a_roi      =  int(a_roi_str[0])

    
#initializing month and total savings variable.
month         = 0
total_savings = 0

#calculating number of months required to afford downpayment
while total_savings < downpayment:
    total_savings = total_savings + (a_salary*m_save_perc/1200) + (total_savings*a_roi/1200)
    month         = month + 1
    
    # salary increases every six months. hence multiplying annualy salary by semi annual raise 
    if month % 6 == 0:
        a_salary  = a_salary + a_salary*(sa_salary_raise/100)
        continue

print()

print('You would have to work for %d months to afford the downpayment for your dream house'%month)
print('%d months or %d years and %d months.'%(month,month//12,month%12))

print()

#calculate number of months from the given years to use it in calculating best percentage later.
years  = int(input("How many years are you planning to save for downpayment? "))
months = years*12


#initializing variable for saving current savings.
current_savings = 0.0


# used for setting error margin in the down payment and savings possible using best percentage of monthly salary
# total savings can be more than down payment or less than payment by this value
epsilon = 100

# using bisection search logic to set high and low value for percentage of monthly salary 
high_initial = 10000
high         = high_initial
low          = 0


# with the use of  bisection search logic, using the middle value as best percentage of monthly salary for all the iterations. 
m_save_perc_best = (high+low)//2


#to calculate number of iterations.
steps = 0

#calculating the best percentage of monthly salary till the error margin is greater than 100. 
while abs(current_savings - downpayment) > epsilon:
    
    steps           = steps + 1
    current_savings = 0.0
    a_salary        = a_salary_base
    m_salary        = a_salary/12
    m_savings       = m_salary*m_save_perc_best/10000
    
    #calculating savings for number of months in given years with the monthly saving percentage for current iteration
    for month in range (1,months+1):
        current_savings += current_savings*a_roi/1200
        current_savings += m_savings

        # making sure we consider semi annual raise in the salary        
        if month % 6 == 0:
            a_salary += a_salary*sa_salary_raise/100
            m_salary  = a_salary/12
            m_savings = m_salary*m_save_perc_best/10000
            
            
    # save the value of m_save_perc_best  in another variable so that it can be used for 
    previous_save_perc_best = m_save_perc_best
    
    
    # check if the total savings using the monthly saving percentage on current iteration is greater than or is less than downpayment
    # if current saving is greater than downpayment, we need to select lesser value for m_save_perc_best in the next iteration
    # if current saving is less than downpayment, we need to select greater value for m_save_perc_best in the next iterations

    if current_savings > downpayment:
        high = previous_save_perc_best
    else:
        low  = m_save_perc_best
        
        
        
        
    # setting the value of m_save_perc_best as mid point of high and low value. 
    # values for high and low have been already set in the previous set based on the current savings. 
    
    m_save_perc_best = int(round((high+low)/2))
    
    
    
if previous_save_perc_best == m_save_perc_best and m_save_perc_best == high_initial:
    print('It is not possible to pay the down payment in %d years'%years)
else:
    print('The best savings rate for you is %.4f' %(m_save_perc_best/ 10000))
````
